```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Đọc dữ liệu (giả định bạn đã đọc vào biến df)
df = pd.read_csv("/content/drive/MyDrive/household_power_consumption.csv")  # thay bằng đường dẫn thực tế

# Tiền xử lý
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce')
df = df.dropna(subset=['Datetime'])
df.set_index('Datetime', inplace=True)

# Chuyển đổi sang numeric
for col in ['Global_active_power']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Loại bỏ các giá trị NaN
df = df[['Global_active_power']].dropna()

# Resample theo giờ nếu cần (nếu dữ liệu gốc là phút)
df_hourly = df.resample('H').mean()

# ----- Bước 2: FFT -----
signal = df_hourly['Global_active_power'].values
n = len(signal)
timestep = 1  # 1 giờ
fft_values = fft(signal)
frequencies = fftfreq(n, d=timestep)

# Chỉ lấy phần dương
pos_mask = frequencies > 0
fft_values = np.abs(fft_values[pos_mask])
frequencies = frequencies[pos_mask]

# ----- Bước 3: Xác định chu kỳ -----
# Tìm tần số có biên độ lớn nhất (chu kỳ chiếm ưu thế)
dominant_freq = frequencies[np.argmax(fft_values)]
dominant_period_hours = 1 / dominant_freq
print(f"Chu kỳ chiếm ưu thế: {dominant_period_hours:.2f} giờ")

# Vẽ phổ tần số
plt.figure(figsize=(10, 4))
plt.plot(frequencies, fft_values)
plt.title("Phổ tần số FFT của Global Active Power")
plt.xlabel("Tần số (1/giờ)")
plt.ylabel("Biên độ")
plt.grid(True)
plt.show()

# ----- Bước 4: Trung bình hóa theo giờ trong ngày -----
df_hourly['Hour'] = df_hourly.index.hour
hourly_avg = df_hourly.groupby('Hour')['Global_active_power'].mean()

plt.figure(figsize=(10, 4))
hourly_avg.plot(kind='bar')
plt.title("Mức tiêu thụ trung bình theo giờ trong ngày")
plt.xlabel("Giờ trong ngày")
plt.ylabel("Công suất hoạt động trung bình (kW)")
plt.grid(True)
plt.show()

# Dự đoán khoảng thời gian tiêu thụ cao nhất
peak_hour = hourly_avg.idxmax()
print(f"Khoảng thời gian tiêu thụ cao nhất là vào {peak_hour}:00")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.signal import find_peaks
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Thiết lập font tiếng Việt và style
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class PowerConsumptionAnalyzer:
    def __init__(self, data_path=None, df=None):
        """
        Khởi tạo analyzer
        Args:
            data_path: đường dẫn tới file dữ liệu
            df: DataFrame nếu đã có sẵn
        """
        if df is not None:
            self.df = df.copy()
        elif data_path:
            self.df = pd.read_csv(data_path)
        else:
            raise ValueError("Cần cung cấp data_path hoặc DataFrame")

        self.prepare_data()

    def prepare_data(self):
        """Chuẩn bị và làm sạch dữ liệu"""
        print("Đang chuẩn bị dữ liệu...")

        # Kết hợp Date và Time thành DateTime
        self.df['DateTime'] = pd.to_datetime(self.df['Date'] + ' ' + self.df['Time'])

        # Chuyển đổi các cột object thành numeric, thay '?' bằng NaN
        numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage',
                       'Global_intensity', 'Sub_metering_1', 'Sub_metering_2']

        for col in numeric_cols:
            self.df[col] = pd.to_numeric(self.df[col].replace('?', np.nan), errors='coerce')

        # Sắp xếp theo thời gian
        self.df = self.df.sort_values('DateTime').reset_index(drop=True)

        # Tạo cột tổng công suất tiêu thụ (Global_active_power là chỉ số chính)
        self.df['Total_Power'] = self.df['Global_active_power'].fillna(method='ffill')

        # Loại bỏ các giá trị NaN còn lại
        self.df = self.df.dropna(subset=['Total_Power'])

        print(f"Dữ liệu sau khi xử lý: {len(self.df)} mẫu")
        print(f"Khoảng thời gian: {self.df['DateTime'].min()} đến {self.df['DateTime'].max()}")

    def resample_hourly(self):
        """Resample dữ liệu theo giờ nếu cần"""
        # Kiểm tra tần suất dữ liệu hiện tại
        time_diff = self.df['DateTime'].diff().mode()[0]
        print(f"Tần suất dữ liệu hiện tại: {time_diff}")

        # Nếu dữ liệu không phải theo giờ, resample về theo giờ
        if time_diff != pd.Timedelta(hours=1):
            self.df_hourly = self.df.set_index('DateTime').resample('H')['Total_Power'].mean().reset_index()
        else:
            self.df_hourly = self.df[['DateTime', 'Total_Power']].copy()

        print(f"Dữ liệu hourly: {len(self.df_hourly)} mẫu")
        return self.df_hourly

    def perform_fft_analysis(self):
        """Thực hiện phân tích FFT"""
        print("\nĐang thực hiện phân tích FFT...")

        # Resample dữ liệu theo giờ
        hourly_data = self.resample_hourly()

        # Lấy chuỗi thời gian công suất
        power_series = hourly_data['Total_Power'].values

        # Loại bỏ trend bằng cách trừ moving average
        window = min(24, len(power_series) // 10)  # Window size tối đa 24h
        power_detrended = power_series - pd.Series(power_series).rolling(window=window, center=True).mean().fillna(method='bfill').fillna(method='ffill')

        # Thực hiện FFT
        N = len(power_detrended)
        fft_values = fft(power_detrended)
        fft_magnitude = np.abs(fft_values)

        # Tính tần số (cycles per hour)
        sample_rate = 1  # 1 sample per hour
        frequencies = fftfreq(N, d=1/sample_rate)

        # Chỉ lấy phần dương của phổ
        positive_freq_mask = frequencies > 0
        frequencies_pos = frequencies[positive_freq_mask]
        magnitude_pos = fft_magnitude[positive_freq_mask]

        # Chuyển đổi sang chu kỳ (giờ)
        periods = 1 / frequencies_pos

        self.fft_results = {
            'frequencies': frequencies_pos,
            'magnitudes': magnitude_pos,
            'periods': periods,
            'power_original': power_series,
            'power_detrended': power_detrended,
            'datetime': hourly_data['DateTime']
        }

        return self.fft_results

    def identify_dominant_cycles(self, min_period=2, max_period=168):
        """
        Xác định các chu kỳ chiếm ưu thế
        Args:
            min_period: chu kỳ tối thiểu (giờ)
            max_period: chu kỳ tối đa (giờ) - mặc định 1 tuần
        """
        print(f"\nXác định chu kỳ chiếm ưu thế (từ {min_period}h đến {max_period}h)...")

        # Lọc theo khoảng chu kỳ quan tâm
        mask = (self.fft_results['periods'] >= min_period) & (self.fft_results['periods'] <= max_period)
        periods_filtered = self.fft_results['periods'][mask]
        magnitudes_filtered = self.fft_results['magnitudes'][mask]

        # Tìm peaks trong phổ magnitude
        peaks, properties = find_peaks(magnitudes_filtered,
                                     height=np.percentile(magnitudes_filtered, 90),  # Chỉ lấy peaks cao
                                     distance=len(magnitudes_filtered)//20)  # Tránh peaks quá gần nhau

        # Lấy các chu kỳ quan trọng nhất
        dominant_periods = periods_filtered[peaks]
        dominant_magnitudes = magnitudes_filtered[peaks]

        # Sắp xếp theo magnitude giảm dần
        sorted_indices = np.argsort(dominant_magnitudes)[::-1]
        dominant_periods = dominant_periods[sorted_indices]
        dominant_magnitudes = dominant_magnitudes[sorted_indices]

        self.dominant_cycles = []

        print("\nCác chu kỳ chiếm ưu thế:")
        for i, (period, magnitude) in enumerate(zip(dominant_periods[:10], dominant_magnitudes[:10])):
            cycle_type = self.classify_cycle(period)
            self.dominant_cycles.append({
                'period_hours': period,
                'magnitude': magnitude,
                'type': cycle_type
            })
            print(f"{i+1}. Chu kỳ: {period:.1f} giờ ({period/24:.1f} ngày) - {cycle_type} - Magnitude: {magnitude:.0f}")

        return self.dominant_cycles

    def classify_cycle(self, period_hours):
        """Phân loại chu kỳ dựa trên thời gian"""
        if 20 <= period_hours <= 28:
            return "Chu kỳ hàng ngày (24h)"
        elif 160 <= period_hours <= 176:
            return "Chu kỳ hàng tuần (7 ngày)"
        elif 10 <= period_hours <= 14:
            return "Chu kỳ nửa ngày (12h)"
        elif 6 <= period_hours <= 10:
            return "Chu kỳ 8 giờ"
        elif 4 <= period_hours <= 6:
            return "Chu kỳ 4-6 giờ"
        else:
            return f"Chu kỳ {period_hours:.1f} giờ"

    def predict_peak_hours(self):
        """Dự đoán khoảng thời gian tiêu thụ cao trong ngày"""
        print("\nPhân tích pattern tiêu thụ theo giờ trong ngày...")

        # Tạo DataFrame với giờ trong ngày
        hourly_data = self.resample_hourly()
        hourly_data['Hour'] = hourly_data['DateTime'].dt.hour

        # Tính tiêu thụ trung bình theo giờ
        hourly_avg = hourly_data.groupby('Hour')['Total_Power'].agg(['mean', 'std']).reset_index()

        # Xác định giờ cao điểm (trên percentile 75)
        threshold = np.percentile(hourly_avg['mean'], 75)
        peak_hours = hourly_avg[hourly_avg['mean'] >= threshold]['Hour'].tolist()

        # Tìm các khoảng thời gian liên tiếp
        peak_periods = []
        if peak_hours:
            start = peak_hours[0]
            end = peak_hours[0]

            for hour in peak_hours[1:]:
                if hour == end + 1:
                    end = hour
                else:
                    peak_periods.append(f"{start:02d}:00 - {end:02d}:59")
                    start = hour
                    end = hour
            peak_periods.append(f"{start:02d}:00 - {end:02d}:59")

        self.hourly_pattern = {
            'hourly_avg': hourly_avg,
            'peak_hours': peak_hours,
            'peak_periods': peak_periods,
            'threshold': threshold
        }

        print(f"Ngưỡng tiêu thụ cao: {threshold:.2f} kW")
        print("Các khoảng thời gian tiêu thụ cao trong ngày:")
        for period in peak_periods:
            print(f"  - {period}")

        return self.hourly_pattern

    def plot_comprehensive_analysis(self):
        """Vẽ biểu đồ tổng hợp phân tích"""
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle('Phân Tích Phổ Fourier - Tiêu Thụ Điện Năng', fontsize=16, fontweight='bold')

        # 1. Chuỗi thời gian gốc
        ax1 = axes[0, 0]
        sample_data = self.df.iloc[::24]  # Lấy mẫu mỗi 24 giờ để tránh quá tải
        ax1.plot(sample_data['DateTime'], sample_data['Total_Power'], alpha=0.7, linewidth=0.5)
        ax1.set_title('Chuỗi Thời Gian Tiêu Thụ Điện Năng')
        ax1.set_xlabel('Thời gian')
        ax1.set_ylabel('Công suất (kW)')
        ax1.tick_params(axis='x', rotation=45)
        ax1.grid(True, alpha=0.3)

        # 2. Phổ tần số FFT
        ax2 = axes[0, 1]
        # Chỉ hiển thị chu kỳ từ 1 giờ đến 7 ngày
        mask = (self.fft_results['periods'] >= 1) & (self.fft_results['periods'] <= 168)
        periods_plot = self.fft_results['periods'][mask]
        magnitudes_plot = self.fft_results['magnitudes'][mask]

        ax2.loglog(periods_plot, magnitudes_plot, alpha=0.7)
        ax2.set_title('Phổ Tần Số FFT')
        ax2.set_xlabel('Chu kỳ (giờ)')
        ax2.set_ylabel('Magnitude')
        ax2.grid(True, alpha=0.3)

        # Đánh dấu các chu kỳ quan trọng
        for cycle in self.dominant_cycles[:5]:
            ax2.axvline(x=cycle['period_hours'], color='red', linestyle='--', alpha=0.7)
            ax2.text(cycle['period_hours'], cycle['magnitude'],
                    f"{cycle['period_hours']:.1f}h", rotation=90,
                    verticalalignment='bottom', fontsize=8)

        # 3. Pattern theo giờ trong ngày
        ax3 = axes[1, 0]
        hourly_avg = self.hourly_pattern['hourly_avg']
        bars = ax3.bar(hourly_avg['Hour'], hourly_avg['mean'],
                      color=['red' if h in self.hourly_pattern['peak_hours'] else 'blue'
                             for h in hourly_avg['Hour']], alpha=0.7)
        ax3.errorbar(hourly_avg['Hour'], hourly_avg['mean'],
                    yerr=hourly_avg['std'], fmt='none', color='black', alpha=0.5)
        ax3.set_title('Pattern Tiêu Thụ Theo Giờ Trong Ngày')
        ax3.set_xlabel('Giờ trong ngày')
        ax3.set_ylabel('Tiêu thụ trung bình (kW)')
        ax3.axhline(y=self.hourly_pattern['threshold'], color='red', linestyle='--', alpha=0.7, label='Ngưỡng cao điểm')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        ax3.set_xticks(range(0, 24, 2))

        # 4. Magnitude của các chu kỳ chính
        ax4 = axes[1, 1]
        top_cycles = self.dominant_cycles[:8]
        cycle_labels = [f"{c['period_hours']:.1f}h\n({c['type'].split('(')[0]})"
                       for c in top_cycles]
        magnitudes = [c['magnitude'] for c in top_cycles]

        bars = ax4.bar(range(len(top_cycles)), magnitudes, alpha=0.8,
                      color=plt.cm.viridis(np.linspace(0, 1, len(top_cycles))))
        ax4.set_title('Magnitude Các Chu Kỳ Chính')
        ax4.set_xlabel('Chu kỳ')
        ax4.set_ylabel('Magnitude')
        ax4.set_xticks(range(len(top_cycles)))
        ax4.set_xticklabels(cycle_labels, rotation=45, ha='right')
        ax4.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def generate_summary_report(self):
        """Tạo báo cáo tổng kết"""
        print("\n" + "="*60)
        print("BÁO CÁO PHÂN TÍCH CHU KỲ TIÊU THỤ ĐIỆN NĂNG")
        print("="*60)

        print(f"\n📊 THÔNG TIN DỮ LIỆU:")
        print(f"  • Số mẫu: {len(self.df):,}")
        print(f"  • Khoảng thời gian: {self.df['DateTime'].min()} → {self.df['DateTime'].max()}")
        print(f"  • Tiêu thụ trung bình: {self.df['Total_Power'].mean():.2f} kW")
        print(f"  • Tiêu thụ tối đa: {self.df['Total_Power'].max():.2f} kW")

        print(f"\n🔄 CHU KỲ CHIẾM ƯU THẾ:")
        for i, cycle in enumerate(self.dominant_cycles[:5], 1):
            print(f"  {i}. {cycle['type']}")
            print(f"     - Chu kỳ: {cycle['period_hours']:.1f} giờ")
            print(f"     - Độ mạnh: {cycle['magnitude']:.0f}")

        print(f"\n⏰ KHOẢNG THỜI GIAN TIÊU THỤ CAO:")
        for period in self.hourly_pattern['peak_periods']:
            print(f"  • {period}")

        print(f"\n💡 KẾT LUẬN VÀ KHUYẾN NGHỊ:")

        # Tìm chu kỳ 24h
        daily_cycle = next((c for c in self.dominant_cycles if 20 <= c['period_hours'] <= 28), None)
        if daily_cycle:
            print(f"  ✓ Có chu kỳ hàng ngày rõ ràng ({daily_cycle['period_hours']:.1f}h)")
        else:
            print(f"  ⚠ Chu kỳ hàng ngày không rõ ràng")

        # Tìm chu kỳ tuần
        weekly_cycle = next((c for c in self.dominant_cycles if 160 <= c['period_hours'] <= 176), None)
        if weekly_cycle:
            print(f"  ✓ Có chu kỳ hàng tuần ({weekly_cycle['period_hours']:.1f}h = {weekly_cycle['period_hours']/24:.1f} ngày)")
        else:
            print(f"  ⚠ Chu kỳ hàng tuần không rõ ràng")

        peak_hours_count = len(self.hourly_pattern['peak_hours'])
        print(f"  • Có {peak_hours_count} giờ cao điểm trong ngày")

        if peak_hours_count <= 6:
            print(f"  💡 Khuyến nghị: Tối ưu hóa sử dụng điện trong các giờ cao điểm")
        else:
            print(f"  💡 Khuyến nghị: Xem xét phân tán tải trong ngày")

# Hướng dẫn sử dụng
def main():
    """
    Hướng dẫn sử dụng analyzer
    """
    print("=== HƯỚNG DẪN SỬ DỤNG ===")
    print("1. Khởi tạo analyzer với DataFrame hoặc file path:")
    print("   analyzer = PowerConsumptionAnalyzer(df=your_dataframe)")
    print("   # hoặc")
    print("   analyzer = PowerConsumptionAnalyzer(data_path='path_to_your_file.csv')")
    print()
    print("2. Thực hiện phân tích:")
    print("   analyzer.perform_fft_analysis()")
    print("   analyzer.identify_dominant_cycles()")
    print("   analyzer.predict_peak_hours()")
    print()
    print("3. Vẽ biểu đồ và tạo báo cáo:")
    print("   analyzer.plot_comprehensive_analysis()")
    print("   analyzer.generate_summary_report()")
    print()
    print("=== VÍ DỤ ===")
    print("""
# Giả sử bạn có DataFrame tên là 'df'
analyzer = PowerConsumptionAnalyzer(df=df)

# Thực hiện phân tích đầy đủ
fft_results = analyzer.perform_fft_analysis()
dominant_cycles = analyzer.identify_dominant_cycles()
peak_hours = analyzer.predict_peak_hours()

# Hiển thị kết quả
analyzer.plot_comprehensive_analysis()
analyzer.generate_summary_report()
    """)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Đọc dữ liệu (giả sử đã đọc thành df)
# df = pd.read_csv('file.csv')  # nếu cần

# 1. Tiền xử lý dữ liệu
# df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S') # Original line
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce') # Allow pandas to infer the format
df.set_index('Datetime', inplace=True)

# Chuyển đổi về số
cols_to_float = [
    'Global_active_power', 'Global_reactive_power', 'Voltage',
    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2'
]
for col in cols_to_float:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Lấy chuỗi thời gian theo giờ
df_hourly = df['Global_active_power'].resample('H').mean().dropna()

# 2. Thực hiện FFT
n = len(df_hourly)
fft_vals = np.fft.fft(df_hourly)
fft_freqs = np.fft.fftfreq(n, d=1)  # d=1 vì khoảng cách giữa các điểm là 1 giờ

# Lấy phổ dương (1/2 đầu)
half_n = n // 2
fft_vals = fft_vals[:half_n]
fft_freqs = fft_freqs[:half_n]
amplitudes = np.abs(fft_vals)

# 3. Vẽ phổ tần số
plt.figure(figsize=(12, 5))
plt.plot(fft_freqs, amplitudes)
plt.title("Phổ tần số tiêu thụ điện năng")
plt.xlabel("Tần số (1/giờ)")
plt.ylabel("Biên độ")
plt.grid()
plt.show()

# 4. Tìm chu kỳ chiếm ưu thế
dominant_idx = np.argmax(amplitudes[1:]) + 1  # bỏ thành phần DC (tần số 0)
dominant_freq = fft_freqs[dominant_idx]
dominant_period_hours = 1 / dominant_freq
print(f"Chu kỳ chiếm ưu thế: {dominant_period_hours:.2f} giờ (~ {dominant_period_hours / 24:.2f} ngày)")

# 5. Xác định khoảng tiêu thụ cao trong ngày
df_hourly_by_hour = df_hourly.groupby(df_hourly.index.hour).mean()

plt.figure(figsize=(10, 4))
plt.plot(df_hourly_by_hour.index, df_hourly_by_hour.values, marker='o')
plt.title("Trung bình tiêu thụ điện năng theo giờ trong ngày")
plt.xlabel("Giờ trong ngày")
plt.ylabel("Công suất trung bình (kW)")
plt.grid()
plt.xticks(range(0, 24))
plt.show()

peak_hour = df_hourly_by_hour.idxmax()
print(f"Giờ tiêu thụ cao nhất trong ngày: {peak_hour}:00")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Đọc dữ liệu (giả sử đã đọc thành df)
df = pd.read_csv('/content/drive/MyDrive/household_power_consumption.csv')  # nếu cần

# 1. Tiền xử lý dữ liệu
# df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H:%M:%S') # Original line
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce') # Allow pandas to infer the format
df.set_index('Datetime', inplace=True)

# Chuyển đổi về số
cols_to_float = [
    'Global_active_power', 'Global_reactive_power', 'Voltage',
    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2'
]
for col in cols_to_float:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Lấy chuỗi thời gian theo giờ
df_hourly = df['Global_active_power'].resample('H').mean().dropna()

# 2. Thực hiện FFT
n = len(df_hourly)
fft_vals = np.fft.fft(df_hourly)
fft_freqs = np.fft.fftfreq(n, d=1)  # d=1 vì khoảng cách giữa các điểm là 1 giờ

# Lấy phổ dương (1/2 đầu)
half_n = n
fft_vals = fft_vals[:half_n]
fft_freqs = fft_freqs[:half_n]
amplitudes = np.abs(fft_vals)

# 3. Vẽ phổ tần số
plt.figure(figsize=(12, 5))
plt.plot(fft_freqs, amplitudes)
plt.title("Phổ tần số tiêu thụ điện năng")
plt.xlabel("Tần số (1/giờ)")
plt.ylabel("Biên độ")
plt.grid()
plt.show()

# 4. Tìm chu kỳ chiếm ưu thế
dominant_idx = np.argmax(amplitudes[1:]) + 1  # bỏ thành phần DC (tần số 0)
dominant_freq = fft_freqs[dominant_idx]
dominant_period_hours = 1 / dominant_freq
print(f"Chu kỳ chiếm ưu thế: {dominant_period_hours:.2f} giờ (~ {dominant_period_hours / 24:.2f} ngày)")

# 5. Xác định khoảng tiêu thụ cao trong ngày
df_hourly_by_hour = df_hourly.groupby(df_hourly.index.hour).mean()

plt.figure(figsize=(10, 4))
plt.plot(df_hourly_by_hour.index, df_hourly_by_hour.values, marker='o')
plt.title("Trung bình tiêu thụ điện năng theo giờ trong ngày")
plt.xlabel("Giờ trong ngày")
plt.ylabel("Công suất trung bình (kW)")
plt.grid()
plt.xticks(range(0, 24))
plt.show()

peak_hour = df_hourly_by_hour.idxmax()
print(f"Giờ tiêu thụ cao nhất trong ngày: {peak_hour}:00")
```

```python
import pandas as pd
import numpy as np

# Giả sử bạn đã đọc DataFrame là df
# df = pd.read_csv(...)

# Chuyển đổi DatetimeIndex (nếu chưa làm)
df=dff

# Chuyển cột 'Global_active_power' về dạng số (phòng trường hợp có chuỗi rỗng)
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')

# Kiểm tra và xử lý missing values
df = df[['Global_active_power']].copy()
df = df.fillna(method='ffill')  # forward fill
```

```python
df.sort_index(inplace=True)
```

```python
train = df.loc[:'2007-05-31 00:00:00']
test = df.loc['2007-06-01 00:00:00':]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train)

# Hàm tạo dữ liệu chuỗi thời gian
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

window_size = 60  # dùng 60 phút quá khứ để dự đoán
X_train, y_train = create_sequences(train_scaled, window_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=10, batch_size=64)
```

```python
# Chuẩn bị dữ liệu test tương tự
test_scaled = scaler.transform(test)
X_test, y_test = create_sequences(test_scaled, window_size)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Dự đoán
y_pred_scaled = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_true = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(y_true, label='Thực tế', color='blue')
plt.plot(y_pred, label='Dự đoán', color='red')
plt.title('Dự đoán tiêu thụ điện năng - LSTM')
plt.xlabel('Thời gian (phút)')
plt.ylabel('Công suất hoạt động (kW)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lấy chỉ số thời gian tương ứng với các dự đoán
pred_index = test.index[window_size:]

# Tạo DataFrame chứa dự đoán và thời gian
df_pred = pd.DataFrame({
    'datetime': pred_index,
    'y_pred': y_pred.flatten()
})
df_pred.set_index('datetime', inplace=True)
```

```python
# Trích xuất giờ từ datetime
df_pred['hour'] = df_pred.index.hour

# Tính trung bình dự đoán theo mỗi giờ
hourly_avg = df_pred.groupby('hour')['y_pred'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(hourly_avg.index, hourly_avg.values, marker='o', color='green')
plt.title('Trung bình điện tiêu thụ theo giờ trong ngày (Dự đoán)')
plt.xlabel('Giờ trong ngày (0-23)')
plt.ylabel('Công suất tiêu thụ trung bình (kW)')
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```
